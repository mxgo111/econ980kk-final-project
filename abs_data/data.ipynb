```python
df = pd.read_csv("2017-2019-aggregated-cleaned.csv")
df["Product Innovation"] /= 100
df["Business Process Innovation"] /= 100
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x="Product Innovation", label="Product Innovation", stat="percent", bins=15, kde=True)
sns.histplot(data=df, x="Business Process Innovation", label="Business Process Innovation", stat="percent", bins=15, kde=True)
plt.legend()
plt.title("Distribution of Product and Business Process Innovation Probabilities")
plt.xlabel("Probability of Innovation in Three Years")
plt.ylabel("Percent of Sectors")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df["Business Process / Product Innovation Ratio"] = df["Business Process Innovation"] / df["Product Innovation"]
sns.histplot(data=df, x="Business Process / Product Innovation Ratio", stat="percent", bins=15, color="green", kde=True)
plt.title("Distribution of (Business Process / Product) Innovation Ratios")
plt.xlabel("Ratio of Business Process Innovation Rate to Product Innovation Rate")
plt.ylabel("Percent of Sectors")
plt.show()
```
